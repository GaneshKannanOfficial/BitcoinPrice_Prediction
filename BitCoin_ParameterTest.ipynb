```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller

ticker_symbol = "BTC-USD"
start_date = "2010-01-01"
end_date = "2023-01-01"
btc_data = yf.download(ticker_symbol, start=start_date, end=end_date, progress=False)

btc_adj_close = btc_data['Adj Close']

plt.figure(figsize=(12, 6))
plt.subplot(211)
plot_acf(btc_adj_close, ax=plt.gca(), lags=40)
plt.subplot(212)
plot_pacf(btc_adj_close, ax=plt.gca(), lags=40)
plt.show()

adf_result = adfuller(btc_adj_close)
adf_statistic = adf_result[0]
adf_pvalue = adf_result[1]

print(f'ADF Statistic: {adf_statistic}')
print(f'p-value: {adf_pvalue}')

d = 0
if adf_pvalue > 0.05:
    d = 1

print(f'Order of differencing (d): {d}')
```
